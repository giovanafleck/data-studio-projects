# Amazon Fund

Jair Bolsonaro’s policy regarding the environment is causing concern all over the world. In the past 10 years, the Amazon Fund - which is based on investments from Norway and Germany - has helped the Brazilian government to reduce deforestation. But both European countries are threatening to end the financing if the current government doesn’t change it’s plans for the future of Amazon (https://www.theguardian.com/world/2019/jan/02/brazil-jair-bolsonaro-amazon-rainforest-protections). In comparison with last year, deforestation has increased 88% in the legal area (http://www.reuters.tv/v/PAA3/2019/07/04/brazil-deforestation-exceeds-88-in-june).
My plan is to scrape the Fund’s website to find: 
How much money was invested in the past 10 years
Where did the money came from 
Where has it been applied 
Did it changed with the passage of years?
